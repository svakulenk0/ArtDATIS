Now let's merge the lists of entities

```python
def merge_ngrams(ngrams_1, ngrams_2):
    ngrams_raw = sorted(ngrams_1 + ngrams_2, key = lambda x: -x[-1])
    cnt = 0
    n = len(ngrams_raw) - 2

    ngrams_all = []
    while cnt < n:
        if ngrams_raw[cnt][0] == ngrams_raw[cnt+1][0]:
            new_item = (ngrams_raw[cnt][0], ngrams_raw[cnt][1] + ngrams_raw[cnt+1][1])
            ngrams_all.append(new_item)
            cnt += 2
        else:
            ngrams_all.append(ngrams_raw[cnt])
            cnt += 1
    return ngrams_all
```

```python
import string
stop_characters = string.punctuation + '–»“’'
def prepare_entities(path_to_entities='data/dutch_entities.p'):
    entities = pickle.load(open(path_to_entities, 'rb'))
    res = []
    for item in entities:
        if item[0][1] != 'CARDINAL' and not item[0][0] in stop_characters:
            res.append(item)
    return res
```

```python
clean_french = prepare_entities('data/french_entities.p')
```

```python
clean_german = prepare_entities('data/german_entities.p')
```

```python
def extract_ngrams(entities):
    bigrams = []
    trigrams = []
    quadrograms = []
    pentagrams = [] 
    hexagrams = []

    for item in entities:
        gram = len(item[0][0].split(' '))
        if gram == 2:
            bigrams.append(item)
        if gram == 3:
            trigrams.append(item)
        if gram == 4:
            quadrograms.append(item)
        if gram == 5:
            pentagrams.append(item)
        if gram == 6: 
            hexagrams.append(item)
            
    return bigrams, trigrams, quadrograms, pentagrams, hexagrams
```

```python
bigrams_french, trigrams_french, quadrograms_french, pentagrams_french, hexagrams_french = extract_ngrams(clean_french)
```

```python
def merge_multiple_ngrams(list_of_langs):
    res = list_of_langs[0]
    cnt = 1
    while cnt < len(list_of_langs):
        res = merge_ngrams(res, list_of_langs[cnt])
        cnt += 1
    return res
```

```python
def write_ngrams(ngrams, filepath='data/ngrams.csv'):
    with open(filepath, 'w') as f:
        f.write('ENTITY, TYPE, COUNT\n')
        for item in ngrams:
            f.write("{0},{1},{2}\n".format(item[0][0].replace('\n', ' '), item[0][1], item[1]))
```
